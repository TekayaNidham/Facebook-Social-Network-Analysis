```python
import pandas as pd
import numpy as np
import networkx as nx
import math
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
import csv
import cv2
from google.colab import files
```

# **Dataset Importing**

---



```python
fb = nx.read_edgelist("facebook_combined.txt")
```

# **Degree centrality**

---



```python
degree_cent_fb = nx.degree_centrality(fb)

print("Facebook degree centrality: ",
    sorted (degree_cent_fb.items(),
            key = lambda x: x[1],
            reverse = True)[:10])

#writing the previous into a csv file 
with open('degree_cent_fb.csv', 'w') as f:
    for key in degree_cent_fb.keys():
        f.write("%s,%s\n"%(key,degree_cent_fb[key]))
#download csv to local
files.download('degree_cent_fb.csv') 
```

```python
# store nodes positions in a dictionnary pos
pos = nx.spring_layout(fb)
# assign a color code to each node
node_color = [20000.0 * fb.degree(v) for v in fb]
# assign size based on value
node_size =  [v * 10000 for v in degree_cent_fb.values()]
# displaying the graph
plt.figure(figsize=(10,10))
nx.draw_networkx(fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
#sorted(degree_cent_fb, values=degree_cent_fb.get, reverse=True)[:10]
```

# **Question 4 : Average degree**

---
According to the networkx [documentation](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.assortativity.average_degree_connectivity.html) Average degree connectivity is refered to as k_nearest_neighbors


```python
def average_degree(g):
  return nx.k_nearest_neighbors(g)
average_degree(fb)
```

```python
fb_n , fb_k = fb.order(), fb.size()
fb_avg_deg = fb_k / fb_n
print("Nodes: ", fb_n)
print("Edges: ", fb_k)
print("Average degree: ", fb_avg_deg)
```
